+ ### **What is recursion ?**

    Recursion is the process a procedure goes through when one of the steps of the procedure involves invoking the procedure itself. A procedure that goes through recursion is said to be 'recursive'.

    From this we are able to make recusrsive function which calls themself to solve the problem 

+ ### **How to approach the problem and solve it with recursion ?**

    Here the concept of PMI(Principle of mathamatical induction) comes into the picture where it tells us which cases are valid and how we can implement to it 

    It is a technique that is used to prove the basic theorems in mathematics which involve the solution up to n finite natural terms. 

    The principle of Mathematical Induction is widely used in proving various statements such as a sum of first n natural numbers is given by the formula ***n(n+1)/2***. This can be easily proved using the Principle of Mathematical Induction.

    Mathematical Induction is one of the fundamental methods of writing proofs and it is used to prove a given statement about any well-organized set. Generally, it is used for proving results or establishing statements that are formulated in terms of n, where n is a natural number. Suppose P(n) is a statement for n natural number then it can be proved using the Principle of Mathematical Induction, Firstly we will prove for P(1) then let P(k) is true then prove for P(k+1). If P(k+1) holds true then we say that P(n) is true by the principle of mathematical induction.

Any statement P(n) which is for “n” natural number can be proved using the Principle of Mathematical Induction by following the below steps,

+ Step 1: Verify if the statement is true for trivial cases (n = 1) i.e. check if P(1) is true.
+ Step 2: Assume that the statement is true for n = k for some k ≥ 1 i.e. P(k) is true.
+ Step 3: If the truth of P(k) implies the truth of P(k + 1), then the statement P(n) is true for all n ≥ 1.

```python
# find factorial

def fact(n):
    if n == 0:
        return 1
    return n*fact(n-1)

n = int(input())
fact(n)
```

```python
# find sum of all natural number

def sum(n):
    if n == 0:
        return 0
    sm = sum(n-1)
    ans = n + sm
    return ans

n = int(input())
sum(n)
```

```python
# find x^n with recursion|

def power(x,n):
    if n == 0:
        return 1
    ans = x * power(x,n-1)
    return ans

x = int(input())
n = int(input())
power(x,n)
```

```python
# print 1 to n and n to 1 number

def print_1_n(n):
    if n == 0:
        return 0
    print_1_n(n-1)
    return print(n , end =" ") 

def print_n_1(n):
    if n == 0:
        return 0
    print(n , end =" ")
    print_n_1(n-1)
    return
    
n = int(input())
print_n_1(n)
print()
print_1_n(n)
```

```python
# fibonacci 

def fibo(n):
    if n == 1 or n == 2:
        return 1
    ans = fibo(n-1)+fibo(n-2)
    return ans

n = int(input())
fibo(n)
```

```python
# find if a number is in array(list) or not 
def find(len, list, x):
    if list[len] == x:
        return True
    elif len == 0:
        if list[0]==x:
            return True
        else:
            return False
    else:
        answer = find(len-1,list,x)
    return answer
n = int(input())
ls =[int(x) for x in input().split()]
x = int(input())

find (n-1,ls,x)
```

```python
# is list sorted or not 
def sort(ls , start):
    n = len(ls)
    if start == n-1 or start == n:
        return True
    if ls[start]>ls[start+1]:
        return False
    ans = sort(ls,start+1)
    return ans

x = [int(x) for x in input().split()]
sort(x,0)
```

```python
# first occurence of x in the list
def findocc(ls, x , a):
    if a == len(ls):
        return -1
    if ls[a] == x:
        return a 
    ans = findocc(ls,x,a+1)
    return ans

ls = [int(x) for x in input().split()]
x = int(input())
findocc(ls,x,0)
```

```python
# find last occurence of the x in list
def findocc(ls, x , a):
    if a == 0 and ls[0]!=x:
        return -1
    if ls[a] == x:
        return a 
    ans = findocc(ls,x,a-1)
    return ans

ls = [int(x) for x in input().split()]
x = int(input())
li = len(ls)
findocc(ls,x,li-1)
```

# **Searching and sorting with recursion**


```python
# binary search
def binser(ls , x , si ,ei):
    if si>ei:
        return -1
    mid = (si+ei)//2
    if ls[mid] == x:
        return mid
    elif x < ls[mid]:
        return binser(ls,x,si,mid-1)
    else: 
        return binser(ls , x ,mid+1 ,ei)

list = [1,2,3,4,5,6,7,14,17,28,89,600,3001]
binser(list, 3001, 0 , len(list))
```

```python
# merge sort

def merge (ls1,ls2,ls):
    i,j,k = 0,0,0
    while i < len(ls1) and j < len(ls2):
        if ls1[i]<ls2[j]:
            ls[k] = ls1[i]
            k+=1
            i+=1
        else :
            ls[k]=ls2[j]
            k+=1
            j+=1
    while i<len(ls1):
        ls[k] = ls1[i]
        i+=1
        k+=1
    while j<len(ls2):
        ls[k]=ls2[j]
        j+=1
        k+=1

def merge_sort(ls):
    if len(ls)==0 or len(ls)==1:
        return 
    mid = len(ls)//2
    ls1 = ls[0:mid]
    ls2 = ls[mid:]
    merge_sort(ls1)
    merge_sort(ls2)
    merge(ls1,ls2,ls)

ls = [int(x) for x in input().split()]
merge_sort(ls)
print(ls)
```

```python
# quick sort

def qui_sor(ls, si ,ei):
    if si>ei:
        return
    pivot_ind= part(ls,si,ei)
    qui_sor(ls,si,pivot_ind-1)
    qui_sor(ls,pivot_ind+1,ei)

def part(ls,si,ei):
    pivot=ls[si]
    count = 0
    for i in range(si,ei+1):
        if ls[i]<pivot:
            count+=1
    ls[si+count] , ls[si] = ls[si],ls[si+count]
    pivot_ind = si+count

    i,j = si,ei
    while i < j:
        if ls[i]<pivot:
            i+=1
        elif ls[j]>=pivot:
            j-=1
        else:
            ls[i],ls[j]=ls[j],ls[i]
            i+=1
            j-=1
    return pivot_ind

ls = [int(x) for x in input().split()]
qui_sor(ls,0,len(ls)-1)
print(ls)
```

```python
# tower of hanoi
def toh(n , a , b , c):
    if n ==1:
        print(f"{n}th disk move from {a} to {c}")
        return
    toh(n-1,a,c,b)
    print(f"{n}th disk move from {a} to {c}")
    toh(n-1, b,a,c)

toh(4,'source','helper','destination')
```
