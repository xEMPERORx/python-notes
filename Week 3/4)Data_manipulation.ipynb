## **Data Manipulation**
Here we will see what type of method we can perform which will help us to make data more clean and how to use that data for future for eda or feature enginering


```python
# lets import the csv to make df out of it 
df = pd.read_csv('./data.csv')
print(df)
```

```python
# print first 5 element with head and last 5 element with tail
# here we can also provide the parameter as the number of rows we need to print from the csv file
print(df.head())
print(df.tail())
```

```python
# describe will give the statistical measures of the data will be there it will not count the categorical data
df.describe()
```

```python
# dtypes will allow us to find the datatype of each column 
df.dtypes
```

```python
## Handling missing values 

# first find the missing value with the help of the isnull function if the bool is false then no missing value over ther and vice versa 
print(df.isnull(),end='\n') # by this we can see there are few places which have missing value lets handle those missing values 
# another way to find it is using any functino 
print(df.isnull().any(),end='\n')
print(df.isnull().any(axis=1),end='\n')

print(df.isnull().sum(),'\n') # this will give the total null values in each column 
```

```python
## how to fill the missing values with differnet value 
# fill the nan value with some value of our choice 
df_filled = df.fillna(0)
df_filled.head(15)
```

```python
# filling missing values with the mean of that cloumn 
df['Sales_filledna'] = df['Sales'].fillna(df['Sales'].mean())
df.head(20)
```

```python
## Renaming a column we have rename functin to do this we need to provde it in the key:value pair to it with old_name:new_name
df.rename(columns={'Date':'Sales Date'},inplace=True)
df.head()
```

```python
## Change the datatype of the column
df['value_new']= df['Value'].fillna(df['Value'].mean()).astype(int)
df.head()
```

```python
## using apply functino which can take function like user or inbuilt or lambda function which will be helped to do opoerations on the column or at row level but preferable to do on column 

df['value_new'] = df['value_new'].apply(lambda x:x*2)
df.head()
```

#### **Data Aggregating and Grouping**

```python
## Data aggreagationg

# group by using one categorical column
grouped_mean = df.groupby('Product')['Value'].mean()
print(grouped_mean)

# with 2 categorical variable 
grouped_sum = df.groupby(['Product','Region'])['Value'].sum()
print(grouped_sum)
grouped_mean = df.groupby(['Product','Region'])['Value'].mean()
print(grouped_mean)

# aggreagte multiple function
grouped_agg = df.groupby('Region')['Value'].agg(['mean','min','sum','count'])
print(grouped_agg)

```

#### **Merging and joining Dataframe**

```python
# Creating DataFrames
left_df = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})

right_df = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K4'],
                      'C': ['C0', 'C1', 'C2', 'C4'],
                      'D': ['D0', 'D1', 'D2', 'D4']})
```

```python
# Merging DataFrames on key column
result = pd.merge(left_df, right_df, on='key')
print(result)

# Different types of merges
# Inner join
result = pd.merge(left_df, right_df, on='key', how='inner')
print(result)

# Outer join
result = pd.merge(left_df, right_df, on='key', how='outer')
print(result)

# left join
result = pd.merge(left_df, right_df, on='key', how='left')
print(result)

# Right join
result = pd.merge(left_df, right_df, on='key', how='right')
print(result)
```
