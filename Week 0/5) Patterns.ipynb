# **simple square pattern in the form of matrix**
if we say print pattern of 3x3<br>
***<br>
***<br>
***<br>
this way we can also make it with number<br>
123<br>
123<br>
123<br>

So lets make this pattern with both while and for loop

```python
#while loop
n = int(input())
i=1
print("in while")
while i<=n:
    j=1
    while j<=n:
        print(j,end='')
        j+=1
    print()
    i+=1
    
print("in for")
#for loop
i=1
for i in range(1,n+1):
    j=1
    for j in range(1,n+1):
        print("*",end='')
    print()
```

# **Triangle patterns**

so here we will talk about how we can print pattern in triangle shape if n is 3 then there can be some type of pattern which we can think of <br>

+ 1<br>

```python
n = int(input())
# first pattern
print("First")
i =1 
while i<=n:
    j=1
    while j<=i:
        print(j,end='')
        j+=1
    print()
    i+=1

# second pattern
print("second")
i = 1 
while i<=n:
    k = i
    j = 1
    while j<=i:
        print(k,end='')
        k+=1
        j+=1
    print()
    i+=1

# third pattern
print("Third")
i = 1 
k = i
while i<=n:
    j = 1
    while j<=i:
        print(k,end='')
        k+=1
        j+=1
    print()
    i+=1
```

# **Character pattern**
So here we will try to print character in here so when we say character we think of a single character so if we have to print ABCD in a single line with a help of a loop we need to add values to the ASCII value of A which eventually give us the value of BCD so first we have to convert that character to its ASCII value in form of int and then we have to convert back to the character itself so for that we have 2 function doing the job.
+ ord('A') - This function will take a single character as input and will give ASCII value as output
+ chr(65) - This function will give the corosponding character to its ASCII value

Types of character pattern 
<br>AAA<br>
BBB<br>
CCC<br>

as we have seen many type of pattern above in those way we can also have same type of character pattern there might be the slight changes in the writing or printing of the code.


```python
#character pattern
n = int(input())
i = 1 
while i <=n:
    j=1
    while j<=n:
        print(chr(ord('A')+i-1),end='')
        j+=1
    print()
    i+=1
```

# **Inverted pattern**
Here the pattern are inverted such as 
<br>****<br>
***<br>
**<br>
*

```python
#inverted pattern
n = int (input())
i = 1
while i<=n:
    j = 1
    while j <= n-i+1:
        print("*",end='')
        j+=1
    print()
    i+=1
```

# **Reversed Pattern**
Here we have to print the pattern but before we were starting with the the start or any other number or character to print so from now onwards we first have to give some space and the we have to print the star after it such as <br>
---?<br>
--??<br>
-???<br>
????<br>

```python
# reversed pattern 
n = int(input())

i = 1 
while i<=n:
    space = 1
    while space <= n-i:
        print(" ",end='')
        space+=1
    star = 1
    while star<= i:
        print("*",end='')
        star+=1
    print()
    i+=1
```

# **Diamond pattern**
pattern such as <br>
--?<br>
-???<br>
?????<br>
-???<br>
--?<br>

```python
n = int(input())

for a1 in range(1, (n+1)//2 + 1): 
    for a2 in range((n+1)//2 - a1):
        print(" ", end = "")
    for a3 in range((a1*2)-1):
        print("*", end = "")
    print()

for a1 in range((n+1)//2 + 1, n + 1):
    for a2 in range(a1 - (n+1)//2):
        print(" ", end = "")
    for a3 in range((n+1 - a1)*2 - 1):
        print("*", end = "")
    print()
```
