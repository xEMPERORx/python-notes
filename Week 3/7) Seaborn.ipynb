# Seaborn

Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

## Installation

```bash
pip install seaborn
```

## Usage

Here is a basic example of how to use Seaborn to create a simple plot:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load an example dataset
tips = sns.load_dataset("tips")

# Create a simple scatter plot
sns.scatterplot(data=tips, x="total_bill", y="tip")

# Show the plot
plt.show()
```

Seaborn provides several built-in datasets and a variety of plot types, including:

- Scatter plots
- Line plots
- Bar plots
- Histograms
- Box plots
- Violin plots
- Heatmaps

You can customize the appearance of your plots using Seaborn's extensive set of options and themes.

````python
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
# start with a normal ploting 
# load data 
tips = sns.load_dataset('tips')

# plot the data as scatter plot
sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title('Scatter plot with Seaborn')
plt.show()
````

````python
## categorical plot

sns.barplot(x='day',y='total_bill',data=tips)
plt.title('Bar plot')
plt.show()
````

````python
## Box Plot
# mainly use when we need to get to know about the outliers
sns.boxplot(x='day',y='total_bill',data=tips)
plt.title('Box plot')
plt.show()
````

````python
## histogram

sns.histplot(tips['total_bill'],bins=10,kde=True) # here kde(kernel destination estimation) is meant for the curve we see 
plt.title('Histogram Plot')
plt.show()
````

````python
# kde plot 
sns.kdeplot(tips['total_bill'],fill=True,color='green')
plt.show()
````

````python
## pair plot
# it is used we want to see the relation between each and every other feature
# Here only the numerical variable are taken
sns.pairplot(tips)
plt.show()
````

````python
## Heatmap
cor = tips[['total_bill','tip','size']].corr()
sns.heatmap(cor,annot=True,cmap='coolwarm')
````
