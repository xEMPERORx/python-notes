### **Else if statement**
When we have a single "if" statement followed by a single "else" statement, if the "if" condition is false, then we go to the "else" statement. The "else" statement does not have any condition attached to it, so if the "if" condition is not true, then we automatically go to the "else" statement.

```python
# find odd and even number taken from user
a = int(input("enter your number"))
if (a%2==0):
    print(a)
    print("it's an even number")
else:
    print(a)
    print("it's an odd number")
```

### **If Elif else statement**
When we need to check multiple conditions at once in a program, we use an if-elif ladder. In this logical structure, the first condition is written in the if statement. If the condition is false, the program moves on to the next condition in the elif statement. Finally, we have the else statement, where we store the default statement to be executed if all the previous conditions are false.

```python
# make a program which checks largest number all three number
a = int(input("enter 1st number : "))
b = int(input("enter 2st number : "))
c = int(input("enter 3st number : "))
print(a,b,c)
if a>b and a>c:
    print(str(a) + " is the largest number ")
elif b>a and b>c:
    print(str(b) + " is the largerst number")
else:
    print(str(c) + " is the largest number")
```

### **Nested statements**
We can use if-else statements in a nested way when we need to check a condition and also have a further condition to verify. For instance, we can check if a condition is true and if it is, we can check one more condition to see if it is also true. This is useful in situations where the first condition being true is a prerequisite for checking the second condition.


```python
# this is just a simple exapmle of nesting of if else
a= True
b= True
if a:
    print("inside a")
    if b:
        print("inside b")
    else:
        print("outside of b")
else:
    print("outside of a")
```

# **Loops**
Loops are similar to if-else statements in that they check a condition, and if the condition is true, they execute the code inside the loop. So, why do we use loops? We use them because they allow us to repeat a certain block of code multiple times, making our code more efficient and less repetitive.
Loops are of 2 types:
+ While
+ for

## **While loop**
The while loop runs a block of code as long as the condition is true. It's important to ensure that the variable used in the condition is updated with each iteration, otherwise the loop may run infinitely.e.

```python
#print number from 1 to n
a = int(input("Enter n : "))
n = 0
while n<=a:
    print(n)
    n+=1
```

```python
#check n is it prime or not

n = int(input("enter n : "))

flag = False
d = 2
while d<n:
    if n%d == 0:
        flag=True
    d+=1
if(flag):
    print("not prime")
else:
    print("Prime")
```

### **Nesting of loop**
As we have seen, we can use nested if statements to revalidate certain conditions. Similarly, if we have two tasks to perform in a loop, we can run an inner loop that executes each iteration of the outer loop. The inner loop will execute completely until the end, within one iteration of the outer loop.

```python
# print all the n prime number
n = int(input("enter n : "))
k = 2

while k<n:
    flag = False
    d = 2
    while d<k:
        if k%d == 0:
            flag=True
        d+=1
    if(not(flag)):
        print(k)
    k+=1
```

# **For loop**
The for loop is a type of loop used in Python to process items one by one from a sequence, such as a list or string.

```python
# print n number using for loop
n = int(input())
for i in range(n):
    print(i)
```

### **range()**
In Python, the range function generates a list, which is a special sequence type that helps to iterate through a for loop.
+ This function have 3 argument range(start,stop,step/stride)
    + Start - The first argument specifies the starting point of the list. The default starting point is 0.
    + Stop - The second argument specifies the endpoint of the list. The list does not include the endpoint it always has stop-1 elements.
    + Step/Stride - The Third argument specifies how much we have to increment/decrement the list with every iteration of the loop. The default step value is 1.
+ Imp Note
    + When we pass only one argument to the range function, it will be considered as the stop argument. This means that the argument will indicate where the list should end. In this case, the start and step values will default to 0 and 1, respectively.
    + When two arguments are passed to the range function, the first argument is considered as the start and the second argument as the stop. The step value will have its default value of 1.

```python
# range with all argument
n = int(input())

for i in range(2,n,3):
    print(i)

#range with 1 argument
print ("1 argument")
for i in range(n):
    print(i)

#range with 2 argument
print("2 argument")
for i in range(1,n+1):
    print(i)
```
