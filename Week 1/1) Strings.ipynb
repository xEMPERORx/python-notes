Strings are one of the most commonly used data types in Python to represent textual data. They are immutable, which means that once a string is created, it cannot be modified. Python also supports various operations on strings such as concatenation, slicing, finding the length of a string, and many more.

### **String slicing**
As we have seen above how the string works so if we need some part of the string then we can use slicing to slice the string from that part and get the sliced string as output.

we can do slice with the help of variable name itself
+ a = "string"
+ a[start:stop:skip]
+ here you have to pass three arguments where start will be the start index of the string and stop will be the till where the slicing should be done, sliced string will not contain the stop index character it is excluded and if you want to get some alternate character then you can input the value in the skip argument.

+ By default start will be 0 and stop will be the end of the string and the skip will be 0 as well


```python
# slicing
a = "emperor"
print(a[2:3])
print(a[:5])
print(a[2:])
print(a[::3])
print(a[-5:-2])
print(a[:])
print(a[6:1:-1])
print(a[6:1:-2])
```

```python
a = 'I am Emperor'
# lower and upper
print(a.upper())
print(a.lower())
print(a.islower())
print(a.isupper())
print(a.isalpha())

#split
print(a.split())
print(a.split('e'))

#replace
print(a.replace('I','you'))
print(a.replace('r','a',1))

#find
print(a.find('I'))
print(a.find('hello'))

#start with
print(a.startswith('I'))
print(a.startswith('you'))
```

### **Docstring**

+ What are the docstrings in Python?
    Python documentation strings (or docstrings) provide a convenient way of associating documentation with Python modules, functions, classes, and methods. It’s specified in source code that is used, like a comment, to document a specific segment of code. Unlike conventional source code comments, the docstring should describe what the function does, not how.

+ Declaring Docstrings: The docstrings are declared using ”’triple single quotes”’ or “”” triple double quotes “”” just below the class, method, or function declaration. All functions should have a docstring.

+ Accessing Docstrings: The docstrings can be accessed using the __doc__ method of the object or using the help function. The below examples demonstrate how to declare and access a docstring.


```python
def doc_str():
    '''This is a docstring example which is showing how a docstring is used this is used to specify what the function can do or can add guideline to how to use the function it works as a function discription'''
    print("this is docstring function")

doc_str()
print(doc_str.__doc__)
```

It was officially written in 2001 by Guido van Rossum, Barry Warsaw, and Nick Coghlan. The main aim of PEP is to enhance the readability and consistency of code.

PEP 8 enhances the readability of the Python code, but why is readability so important? Let's understand this concept.

Creator of Python, Guido van Rossum said, "Code is much more often than it is written." The code can be written in a few minutes, a few hours, or a whole day but once we have written the code, we will never rewrite it again. But sometimes, we need to read the code again and again.

At this point, we must have an idea of why we wrote the particular line in the code. The code should reflect the meaning of each line. That's why readability is so much important.

We will describe few important guidelines for writing effective code that can be read by others as well.
