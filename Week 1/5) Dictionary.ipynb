# **Dictionary**
+ A dictionary is an unordered collection of key-value pairs
+ Indexing in a dictionary is done using these “keys”.
+ Each pair maps the key to its value
+ Literals of the type dictionary are enclosed within curly brackets.
+ Within these brackets, each entry is written as a key followed by a colon ” : ”, which is further followed by a value. This is the value that corresponds to the given key.
+ These keys must be unique and cannot be any immutable data type. Eg- string, integer, tuples, etc. They are always mutable.
+ The values need not be unique. They can be repetitive and can be of any data type (Both mutable and immutable)
+ Dictionaries are mutable, which means that the key-value pairs can be changed.

+ How dicitionary looks
***dictionary = {
    key:value,
    key:value,
    key:value,
    key:value
}***

    

```python
# creating a dictionary
a = {"hi":20 , "hey":30 , 1000:90}
b = dict([(1,2),(2,3),(3,4)])
c = a.copy()
print(a)
print(b)
print(c)

# using an inbulit method
a = dict.fromkeys(["a", "b","this","hey"],2) # this will create all the key to the dictionary with none key value in it  and the 2 here will be the default value for all the dictionary that will only have at most 1 argument
print(a)
```

### **Accessing the element from the dictionary**
Here we have to pass the key value as the index argument which eventually will give the keyvalue of that particular key.


```python
print(a['this'])

# using get method to get the value here if the key is not present it will not give the exception it will give none
print(b.get(3))
print(a.get("this",0)) # here the 0 represents the default value that we have given to it you can give any value to it if the key is present then the key value will get print
```

```python
# some usefull function of the dictionary

print(a.keys()) # this will give all the keys in the dictionary
print(a.values()) # this will give all the values in the dictionary
print(a.items()) # this will give the list of all pairs of key and there values

# checking of key in the dictionary
print('a' in a)

# iterating over the dictionary keys and values
for i in a:
    print(i , a[i])

# iterating over the dictionary values
for i in a.values():
    print(i)
```

## **Add or remove data in dictionary**

we can easily add or remove the element in the dictionary so we can assign a new key with some value it will get added to the dict and if we want to update some vale to a existing key the process reamin same just change the assign value to whatever you want so for delete we can use pop function or we can use clear to clear the whole dictionay and we can also delete an entire dictionart here is how we will be doing

```python
a = {1:2, 'a':56,'gh':'ui','list':[1,2,23]}

# to add new key and value or change the existing value
a['er'] = 'Hi'
a[1] = 'hello'
print(a)

#if we have two dict then we can concate both with the help of update
# here in update the 2nd dict will get concate and the comman content will get update as per 2nd dictionat
b = {1:345345,5:'hi'}
a.update(b)
print(a)

# to remove element fromn the dictionary\
a.pop('er') # this will pop the er named key and its value
print(a)

# clear the dict
a.clear()
print(a)

# delete the dict
del a
print(a)
```

## Merge the dictionary

```python
# merging of two dictionary
a = {1:2,3:4}
b = {5:6,7:8}
merge_dict = {**a,**b}
print(merge_dict)
```
