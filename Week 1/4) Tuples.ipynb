# **Tuples**
+ A tuple is an ordered collection of elements/entries.
+ Objects in a Tuple are immutable i.e. they cannot be altered once created.
+ In a tuple, elements are placed within parentheses, separated by commas.
+ The elements in a tuple can be of different data types like integer, list, etc.
+ Tuple is similar to the list as we will see here most of the things are same as the list like accessing the element and the slicing of tuple and many more fucntion which we use in lists

+ The difference between tuple and list
+ Tuple | List
+ A Tuple is immutable | A list is mutable
+ We cannot change the elements of a Tuple once it is created | We can change the elements of a list once it is created
+ Enclosed within parentheses “( )” | Enclosed within square brackets “[ ]”

+ In tuple we are unable to insert or remove element from it as we have seen they are immutable in nature

```python
# creating a tuple 
a = (1,3,24)
b = 4,5,3
type(b)
```

```python
# Accessing the tuple we can access in positive as well as negative index
print(a[2])
print(b[-1])
```

```python
# slicing in tuple
c = 3,2,3.2,2,1,3,233,3,2
print(c[0:8:2])
```

```python
# concatenantion
d = a+b
print(d)

# for loop on tuple 
for i in d :
    print(i , end = " ")

# len 
print(len(d))

# in 
print(6 in d)

#repetation
print(d*2)

# min and max this will only work when there is comparison is homogenous type
print(min(d))
print(max(d))

# we can also convert list to the tuple
li = [1,2,3,4,54,5]
tup = tuple(li)
print(tup)
```

## **Variable length input and output**
This is somewhat we have to use while using function as we are only able to give a limited amount of parameter to the function but with the help of tuple we are actucally able to give variable length input to a function and can get multiple output in the return statement 

```python
# variable input to the function 

def add(a,b,*input):
    ans = a+b
    for i in input:
        ans +=i
    print(type(input))
    return ans
    
print(add(4,5,2,3,4,5))
print(add(4,5)) # When we pass only 2 parameters to the function the tuple will be empty
print(add(4,5,6))

# Here we have given *input which will be counted as a tuple and store all the values after the default argument
# This way we can get multiple inputs to the function

```

```python
# Variable length output
def add_multi(a,b):
    return a+b,a*b # here the tuple will be generated in the return statement which can be store 
    # on a single value as a tuple or on a multiple variable to store a single statement

d = add_multi(5,6)
print(d)
add,multi = add_multi(6,4)
print(add)
print(multi)
```
