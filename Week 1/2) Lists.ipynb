### **Concatenation , Slicing , Repeating/Replicating element**
The list will have the same concept of concatenation as string we use ***+*** to concate two list with each other the second list will get append to the fist list

Slicing is also done the same way the list name follow with the 3 argument start stop or step we can slice the list in just the same way as we do to the strings.

And as we have seen how a string can be have repeting emelent to it similar list can be also make that with the help of list_name*time to repeate the list

```python
a = [1,2,3,5,60]
b = ['hi','hello','how','are','you']

#concatenation
print(a+b)
print(a+[5,6,214,335841,35131])

#Slicing
print(a[2:4])
print(a[:3])
print(a[1:])
print(a[::-1]) # this is the way which we can directly reverse the list
print(a[-4:-1])

#repeating
print(b*3)
print(a*6)
```

### **Inbuilt function**
+ append():Used for appending/adding elements at the end of a list.
    + Syntax:<ListName>.append(element)


+ extend(): Adds the contents of List2 to the end of List1.
    + Syntax: <ListName1>.extend(<ListName2)
      


+ insert(): Inserts an element at a specified position/index in a list.
    + Syntax: <ListName>(position, element)

+ pop(): It deletes and returns the element at the specified index. If we don't mention the index, it by default pops the last element in the list.

+ del() : Element to be deleted is mentioned using list name and index.

+ remove(): Element to be deleted is mentioned using list name and element.it will remove only the first occurenec of that element in the list.


+ sum() : Returns the sum of all the elements of a List. (Used only for lists containing numerical values)

+ count(): Returns the total number of times a given element occurs in a List.
    + Syntax: <ListName>.count(element)


+ len(): Returns the total length of a List.

+ index(): Returns the index of first occurrence of an element in a list. If element is not present, it returns -1.

+ min() : Returns the minimum element in a List.

+ max(): Returns the maximum element in a List.

Here in above min and max the list should be homogenous as the type of the element should be same otherwise it will give error


```python
li = [1,2,3,4,6,68,46,6,31,32,2]

# append
li.append(56)
print(li)

#insert
li.insert(1,34)
print(li)

#extend
li.extend([555,5664,623,5])
print(li)

#pop
print(li.pop())

#remove
print(li.remove(1))

#delete
print(li)
del li[4]
print(li)

#other func
print(sum(li))
print(li.count(2))
print(len(li))
print(max(li))
print(min(li))
print(li.index(6))
```

### **Taking input for the list**
There are two common ways to take lists as user inputs. These are:-

+ **Line Separated Input of Lists**
+ The way to take line separated input of a list is described below:
    + Create an empty list.
    + Let the number of elements you wish to put in the list be N.
    + Run a loop for N iterations and during these N iterations do the following:
        + Take an element as user input using the input() function.
        + Append this element to the list we created.
    + At the end of N iterations, you would have appended N desired elements to your list.
    + In this, different elements will have to be entered by the user in different lines<br>
+ **Space Separated Input Of List**
+ In Python, a user can take multiple values or inputs in one line by two methods.
    + Using split() method
        + This function helps in taking multiple inputs from the user in a single line . It breaks the given input by the specified separator. If a separator is not provided then any white space is treated as a separator.
        + Syntax:- list_name.split(<sperator>), input.split(<sperator>)  by default sperator will be white spaces
                  
    + Using List comprehension
        + List comprehension is an elegant way to define and create a list in Python. We can create lists just like mathematical statements in one line only.A common syntax to take input using list comprehension is given below.
        + syntax:- inputList= [int(x) for x in input().split()]
        + Basics Syantax [expression for item in iterable]
        + with conditional logic [expression for item in iterable if condition]
        + Nested List Comprehension [expression for item1 in iterable1 for item2 in iterable2]



```python
# line sperated input
li = []
n = int(input())
for i in range(n):
    a = int(input())
    li.append(a)

print(li)
```

```python
# input using split
n = input().split()
print(n)
```

```python
# input using list comprehension

n = [int(a) for a in input().split()]
print(n)
```

```python
# Remove Consecutive Duplicates
n = input()
str = ''
j = 0
for i in n:
    if i!=j:
        str+=i
        j = i
print(str)
```

```python
# reverse each word in the string
n = input().split()
def reverse(n):
    a = len(n)
    str = n[a::-1]
    return str
str = ''
for i in range(len(n)):
    str = str + ' ' + reverse(n[i])

print(str)
```

```python
# Arrange the number in array the number give should be have odd to the left and even to the right

n = int(input())
li1 = []
li2 = []
for i in range(1,n+1):
    if i%2==0:
        li2.append(i)
    else:
        li1.append(i)

print(li1+li2[::-1])
```

```python
# min in list
a = [int(x) for x in input().split()]
m = a[0]
for i in a:
    if(m>i):
        m=i
print('min is {}'.format(m))
```

# **2D Lists**
so here we will se how 2d list are and how they actually work.
+ This is same as the 2d array how we make the array in c and c++
+ We can say that 2d list are list inside list which means in memory they are just stored in just a linear way
+ so when we want to find the element in the 2d list we can use same as the list[i] here in 2d list we have to give both index for row and column list[i][j]
+ Updating the list is the same as in the normal list if we want to change a single element in the row then we need to specify which row and column to change so if we want to change a whole row then just pass the row index and change it with a new single list.


```python
# a 2d list implementation
li =[[1,2,3,4],[3,456,23,1],[3,6,4,5]]
print(li)
```

### **Taking input for the 2d list**
+ here also we can use list comprehension to take input of the list in one line

```python
# input line with space separated
n = int(input())
input= [[int(col) for col in input().split()] for row in range (n)]
print(input)
```

```python
# input with line separated
row = int(input())
column = int(input())
input = input().split()
new_list = [[int(input[i*column+j]) for j in range(column)] for i in range(row)]
print(new_list)
```

It is just same as how we access a 1d list but here we have to specify two parameter of row and column
