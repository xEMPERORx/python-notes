# **Sets**
+ This is an unordered collection of the data in any
+ Mathematically a set is a collection of items (not in any particular order). A Python set is similar to this mathematical definition with below additional conditions.
+ The elements in a set cannot be repeated i.e. an element can occur only once.
+ The elements in the set are immutable(cannot be modified) but the set as a whole is mutable.
+ There is no index attached to any element in a Python set. So they do not support any indexing or slicing operation.
+ The sets in python are typically used for mathematical operations like union, intersection, difference, and complement, etc.


```python
# creating the set
a = {1,2,3,4,5}
b = set() # this is how we can create an empty set
b = {4,5,6,7,8}

#adding element to the set
a.add(10)
print(a)

# removing an element from the set
b.remove(7) # this will give error if the element we want to delete is not in the set
b.discard(6) # this will not give the error if the element is not in the set
# we also have a pop method as well it will give the last value to it and we can use it as well to use for further operations
pop_element = b.pop()
print(pop_element)
print(b)

# functions to perform the set operations
print(a.intersection(b)) # intersection
print(a.union(b)) # union operation
print(a.difference(b)) # here it will have all the element in a which are not in b
print(a.symmetric_difference(b)) # this is union - intersection

# updating function to update the set
a.intersection_update(b)# this will update the a with intersection of a and b
print(a)
a = {1,2,3,4,5}
a.difference_update(b)# this will also update a with the difference
print(a)
a = {1,2,3,4,5}
a.symmetric_difference_update(b) # this will update a with the symmetric difference
print(a)
```

```python
# Set memebership test
print(1 in a) # this will return true
print(1 not in a) # this will return false

a = {1,2,3,4,5}
b = {4,5,6,7,8}
# set subset and superset
print(a.issubset(b)) # this will return false as a is not subset of b
print(a.issuperset(b)) # this will return false as a is not superset of b
print(a.isdisjoint(b)) # this will return false as a and b have
# some common element
a = {1,2,3,4,5}
b = {6,7,8}
print(a.isdisjoint(b)) # this will return true as a and b have no common element
```
